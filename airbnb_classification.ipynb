### 1. Data Preprocessing

#### 1.1 Missing Values

```python
print(df2.columns)
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
# 1. Drop rows with missing target values
df2.dropna(subset=['host_is_superhost'], inplace=True)
```

```python
# 2. Handle missing values in 'host_response_time' by filling with 'Unknown'
df2['host_response_time'].value_counts()
```

```python
df2['host_response_time'].fillna('Unknown', inplace=True)
df2.head()
```

```python
# 3. Handle missing values in 'host_response_rate'
```

```python
# Convert 'host_response_rate' to numeric
df2['host_response_rate'] = df2['host_response_rate'].str.rstrip('%').astype('float') / 100
```

```python
df2['host_response_rate'].describe()
```

```python
df2['host_response_rate_missing'] = df2['host_response_rate'].isna().astype(int)
df2['host_response_rate'] = df2['host_response_rate'].fillna(0)
```

```python
# 4. Handle missing values in 'host_acceptance_rate'
df2['host_acceptance_rate'] = df2['host_acceptance_rate'].str.rstrip('%').astype('float') / 100
df2['host_acceptance_rate_missing'] = df2['host_acceptance_rate'].isna().astype(int)
df2['host_acceptance_rate'] = df2['host_acceptance_rate'].fillna(0)
```

```python
#5. Handle 'host_listings_count' missing values by filling with median
df2['host_listings_count'].describe()
df2['host_listings_count'].fillna(df2['host_listings_count'].median(), inplace=True)
```

```python
#6. Handle 'host_total_listings_count' missing values by filling with median
df2['host_total_listings_count'].describe()
df2['host_total_listings_count'].fillna(df2['host_total_listings_count'].median(), inplace=True)
```

```python
#7. Handle host_has_profile_pic / host_identity_verified missing values by filling with mode
df2['host_has_profile_pic'].value_counts()
df2['host_has_profile_pic'].fillna(df2['host_has_profile_pic'].mode()[0], inplace=True)
df2['host_identity_verified'].fillna(df2['host_identity_verified'].mode()[0], inplace=True)
```

```python
#8. Handle 'bathrooms' missing values with bathrooms_text
# Only get the numeric part from bathrooms_text and convert to float
df2['bathrooms_text'].head(20)
```

```python
# Get the numeric part from bathrooms_text and convert to series
br = df2['bathrooms_text'].str.extract(r'([\d\.]+)').astype(float)
df2['bathrooms'].fillna(br[0], inplace=True)
```

```python
df2['bathrooms'].fillna(0, inplace=True)
```

```python
df2.drop(columns=['bathrooms_text'], inplace=True)
```

```python
#9. Handle 'bedrooms' missing values by filling with median
bed_med = df2['bedrooms'].median()
df2['bedrooms'].fillna(bed_med, inplace=True)
```

```python
#10. Handle prices by creating missing indicator and filling with median
df2['price'] = df2['price'].str.replace('[\$,]', '', regex=True).astype(float)
```

```python
df2['price_missing'] = df2['price'].isna().astype(int)
```

```python
df2.groupby('room_type')['price'].mean()
```

```python
price_pred = df2.groupby(['room_type','accommodates'])['price'].median()
price_pred
```

```python
df2['price'] = df2.groupby(['room_type', 'accommodates'])['price'].transform(lambda x: x.fillna(x.median()))
```

```python
#11. Review missing values
df2['review_rating_missing'] = df2['review_scores_rating'].isna().astype(int)
df2['review_accuracy_missing'] = df2['review_scores_accuracy'].isna().astype(int)
df2['review_clenliness_missing'] = df2['review_scores_cleanliness'].isna().astype(int)
df2['review_checkin_missing'] = df2['review_scores_checkin'].isna().astype(int)
df2['review_communication_missing'] = df2['review_scores_communication'].isna().astype(int)
df2['review_location_missing'] = df2['review_scores_location'].isna().astype(int)
df2['review_value_missing'] = df2['review_scores_value'].isna().astype(int)

df2['review_scores_rating'].fillna(0, inplace=True)
df2['review_scores_accuracy'].fillna(0, inplace=True)
df2['review_scores_cleanliness'].fillna(0, inplace=True)
df2['review_scores_checkin'].fillna(0, inplace=True)
df2['review_scores_communication'].fillna(0, inplace=True)
df2['review_scores_location'].fillna(0, inplace=True)
df2['review_scores_value'].fillna(0, inplace=True)
```

```python
#12. Reviews per month
df2['reviews_per_month_missing'] = df2['reviews_per_month'].isna().astype(int)
df2['reviews_per_month'].fillna(0, inplace=True)
```

```python
#13. Final check
df2.isnull().sum()
```
